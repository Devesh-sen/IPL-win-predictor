```python
import numpy as np
import pandas as pd
```

```python
data1=pd.read_csv("deliveries.csv")
```

```python
match=pd.read_csv("matches.csv")
```

```python
match
```

```python
total_score=data1.groupby(["match_id","inning"])["total_runs"].sum().reset_index()
total_score
```

```python
total_score=total_score[total_score["inning"]==1]
total_score
```

```python
match=match.merge(total_score[["total_runs","match_id"]],left_on="id",right_on="match_id")
match
```

```python
match["team1"].unique()
```

```python
match["team1"]=match["team1"].str.replace('Delhi Daredevils',"Delhi Capitals")
match["team2"]=match["team2"].str.replace('Delhi Daredevils',"Delhi Capitals")
data1["batting_team"]=data1["batting_team"].str.replace('Delhi Daredevils',"Delhi Capitals")
data1["bowling_team"]=data1["bowling_team"].str.replace('Delhi Daredevils',"Delhi Capitals")
```

```python
match["team1"]=match["team1"].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match["team2"]=match["team2"].str.replace('Deccan Chargers','Sunrisers Hyderabad')
data1["batting_team"]=data1["batting_team"].str.replace('Deccan Chargers','Sunrisers Hyderabad')
data1["bowling_team"]=data1["bowling_team"].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match
```

```python
Teams=['Sunrisers Hyderabad', 'Mumbai Indians', 
        'Royal Challengers Bangalore',
       'Kolkata Knight Riders', 'Delhi Capitals', 'Kings XI Punjab',
       'Chennai Super Kings', 'Rajasthan Royals',]
```

```python
match=match[match["team1"].isin(Teams)]
match=match[match["team2"].isin(Teams)]
```

```python
match["team2"].unique()
```

```python
match=match[match["dl_applied"]==0]
```

```python
match
```

```python
match=match[["match_id","city","total_runs","winner"]]
match
```

```python
data1=data1.merge(match,on="match_id")
```

```python
data1=data1[data1["inning"]==2]
data1
```

```python
data1["total_runs_x"].dtype
```

```python
data1["Current_score"]=data1.groupby("match_id")["total_runs_x"].cumsum()
```

```python
data1["Runs_left"]=data1["total_runs_y"]-data1["Current_score"]
```

```python
data1
```

```python
data1["ball_left"]=126-((data1["over"]*6) + (data1["ball"]))
```

```python
data1
```

# wicket left

```python
data1["player_dismissed"]=data1["player_dismissed"].fillna("0")
data1["player_dismissed"]=data1["player_dismissed"].apply(lambda x: x if x=="0" else "1")
data1["player_dismissed"]=data1["player_dismissed"].astype("int")        
```

```python
wicket=data1.groupby("match_id")["player_dismissed"].cumsum().values
```

```python
data1["wicket_left"]=10-wicket
```

```python
data1
```

# Current run rate

```python
data1["Current_rr"]=data1["Current_score"]/((120-data1["ball_left"])/6)
```

# Required run rate

```python
data1["Required_rr"]=data1["Runs_left"]/((data1["ball_left"])/6)
```

```python
data1
```

# winner

```python
def winner(row):
    return 1 if row["batting_team"] == row["winner"] else 0
```

```python
data1["result"]=data1.apply(winner,axis=1)
```

```python
data1
```

```python
data1=data1[data1["ball_left"]!=0]
```

```python
data1
```

```python
data1["batting_team"].unique()
```

# Fetching important col

```python
df=data1[["batting_team","bowling_team","city","Runs_left","ball_left","wicket_left","Current_rr","total_runs_y","Required_rr","result"]]
```

```python
df
```

```python
df=df.sample(frac=1)
```

```python
df.sample()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,x_test,Y_train,y_test= train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
x_test
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf=ColumnTransformer([("trf1",OneHotEncoder(sparse=False,drop="first"),["batting_team","bowling_team","city"])],remainder="passthrough")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

pipe=Pipeline(steps=[("step1",trf),("step2",LogisticRegression())])
```

```python
pipe.fit(X_train,Y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
